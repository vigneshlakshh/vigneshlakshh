```python
#DUMMY variables and one hot encoding
#categorical variables two types
# 1) nominal variables  eg **male,female  **green,blue,red,  **mambalam,kodambakkam,anna nagar
# 2) ordinal variables  eg **phd,masters,bachelors,  **high,medium,low **satisfied,neutral,dissatisfied

#We apply One-Hot Encoding when:The categorical feature is nominal 
#We apply Label Encoding when:The categorical feature is ordinal

# we use one hot encoding to create dummy variables
```

```python
#STEPS
# use pd.get_dummies() function and store in varriable
# use pd.concat() function to concat with original dataframe with the stored values mention axis and store in T2
# drop the categorical variables with names using table.drop([table names]) and mention axis
# drop one dummy varrible column if you have 5 keep only 4 columns
```

```python
import pandas as pd
df=pd.read_csv("F://Machine Learning GITHUB//homeprices5.csv")
df
```

```python
# use pd.get_dummies() function and store in varriable

dummies=pd.get_dummies(df.town)
dummies
```

```python
# use pd.concat() function to concat with original dataframe with the stored values mention axis and store in T2
merged=pd.concat([df,dummies],axis="columns")
merged
```

```python
# drop the categorical variables with names using table.drop([table names]) and mention axis
# drop one dummy varrible column if you have 5 keep only 4 columns
final = merged.drop(['town','west windsor'],axis='columns')
final
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
```

```python
x=final.drop('price',axis="columns")
x
```

```python
y=final.price
y
```

```python
model.fit(x,y)
```

```python
model.predict([[2300,0,1]])
```

```python
model.score(x,y)
```
